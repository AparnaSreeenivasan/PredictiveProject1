This code loads plant leaf images, preprocesses them, and prepares them with labels for training a classification model.

```python
import os
import cv2
import numpy as np

dataset_path = '/content/plantvillage dataset/color'

images = []
labels = []

# Loop through each disease folder
for disease_folder in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, disease_folder)

    # Loop through each image in the folder
    for image_file in os.listdir(folder_path):
        img_path = os.path.join(folder_path, image_file)

        # Read and process image
        img = cv2.imread(img_path)

        if img is not None:
            img = cv2.resize(img, (128, 128))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

            images.append(img)
            labels.append(disease_folder)

# Convert to numpy arrays
images = np.array(images)
labels = np.array(labels)

print("Total images:", len(images))
print("Images shape:", images.shape)
print("Number of classes:", len(np.unique(labels)))
```

Train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
labels = le.fit_transform(labels)

X_train, X_test, y_train, y_test = train_test_split(
    images, labels, test_size=0.2, random_state=42
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```
